```python
img1 = cv2.imread("/content/drive/MyDrive/LGM-Internship/LostInSpace.jpg")

```

Images are represented as rectangular arrays of individually-coloured square pixels, and that the colour of each pixel can be represented as an RGB (Red Blue Green) triplet of numbers. The colours take numerical values from 0 to 255, white being 255 and black being 0. 

```python
img1
```

```python
cv2_imshow(img1)
```

```python
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY) # converts the image to grey scale 
                                                   # uses the formula : Y=0.299*R + 0.587*G + 0.114*B for conversion 
```

```python
cv2_imshow(img1_gray)
```

```python
img1_invert = cv2.bitwise_not(img1_gray)   # INVERTING THE GRAYSCALE, pixel by pixel: (255- image_colour_coordinate)
```

```python
cv2_imshow(img1_invert)
```

Gaussian Blur is used to reduce noise and reduce detail. The image is blurred using a gaussian function. 

```python
img1_blur = cv2.GaussianBlur(img1_invert, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(img1_blur) 

```

Dodging decreases the image exposure while burning increases the exposure.

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

This is essentially dividing the grayscale value of an image pixel A[i] by the inverse of the mask pixel value B[i], while making sure that the resulting pixel value will be in the range [0,255] and that we do not divide by zero

```python
final_img1 = dodgeV2(img1_gray, img1_blur)

cv2_imshow( img1)
cv2_imshow(final_img1)
```
